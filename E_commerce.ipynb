```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv("Sample - Superstore.csv", encoding="latin-1")
```

```python
df.info()
```

### converting data colunms

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
```

```python
df["Ship Date"] = pd.to_datetime(df["Ship Date"])
```

```python
df.head()
```

```python
df["Order Month"] = df["Order Date"].dt.month
df["Order Year"] = df["Order Date"].dt.year
df["Order Day of Week"] = df["Order Date"].dt.dayofweek
```

```python
df.head()
```

### Monthly Sales Analysis

```python
monthly_sales = df.groupby("Order Month")['Sales'].sum().reset_index()
```

```python
monthly_sales
```

```python
fig = px.line(
    monthly_sales,          # Your DataFrame
    x = "Order Month",      # Column for x-axis (time/months)
    y = "Sales",            # Column for y-axis (numeric values)
    title = "Monthly Sales Analysis"  # Chart title
)
fig.show()
```

### Sales by Category

```python
sales_by_category = df.groupby("Category")["Sales"].sum().reset_index()
sales_by_category
```

```python

sales_data = pd.read_csv("Sample - Superstore.csv",encoding="latin-1")

fig = px.pie(sales_data,
             names="Category",
             values="Sales",
             hole=0.2,
             title="Sales Distribution by Category")
fig.show()
```

### Sales Analysis by Sub Category

```python
sales_by_subcategory = df.groupby("Sub-Category")["Sales"].sum().reset_index()
sales_by_subcategory
```

```python
fig = px.bar(sales_by_subcategory, x = "Sub-Category" , y = "Sales", title= "Sales Analysis By Sub Category")
fig.show()
```

### Monthly Profit Analysis

```python
monthly_profit = df.groupby("Order Month")['Profit'].sum().reset_index()
monthly_profit
```

```python
fig = px.line(monthly_profit, x = ("Order Month"), y = ("Profit"), title="Monthly Profit Analysis")
fig.show()
```

### Sales and Profit Analysis by Customer Segment

```python
import plotly.colors as colors
import plotly.graph_objects as go

sales_profit_by_segment = df.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()

color_palette = colors.qualitative.Pastel

fig = go.Figure()
fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'],
                     y=sales_profit_by_segment['Sales'],
                     name='Sales',
                     marker_color=color_palette[0]))

fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'],
                     y=sales_profit_by_segment['Profit'],
                     name='Profit',
                     marker_color=color_palette[1]))

fig.update_layout(title='Sales and Profit Analysis by Customer Segment',
                  xaxis_title='Customer Segment', yaxis_title='Amount')

fig.show()
```


### Analyse sales-to-profit ratio

```python
sales_profit_by_segment = df.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
sales_profit_by_segment['Sales_to_Profit_Ratio'] = sales_profit_by_segment['Sales'] / sales_profit_by_segment['Profit']
print(sales_profit_by_segment[['Segment', 'Sales_to_Profit_Ratio']])
```
